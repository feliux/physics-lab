### Making graphs with matplotlib

```python
import matplotlib.pyplot as plt
plt.style.use("ggplot")

x = np.linspace(-3, 3)
y = np.sin(x)
plt.plot(x, y)
plt.show()
```

#### Plot $x$, $x^2$, $x^3$, $\cdots$ in log-log scale.

```python
def func (x,n):
    y = x**n
    return y

# Create array of x from 10^-2 to 10^2 equally spaced in log scale
x = np.logspace(-2, 2) 

plt.plot(x, func(x, 1), label="$y=x$")
plt.plot(x, func(x, 2), label="$y=x^2$")
plt.plot(x, func(x, 3), label="$y=x^3$")
plt.legend()
plt.xscale("log")
plt.yscale("log")
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.show()
```

```python
x = np.linspace(-np.pi, np.pi)
y = np.sin(x)*np.cos(x)

plt.plot(x,y)
plt.xlabel(r"$\theta$")
plt.legend([r"$\sin\theta\cos\theta$"])
plt.show()
```

#### Histgram plot

```python
np.random.seed(0)
N = 1000
R = np.random.rand(N)

# plot normalized histgram of R using 100 bins
plt.hist(R, bins=100, density=True)
plt.show() # show plot
```

#### Random Walk

```python
NSTEP = 100000
plt.xlabel("number of steps")
plt.ylabel("position")

for nseed in range(10):# generate 10 random walks
    np.random.seed(nseed) 
    # generate random sequencies of NSTEP +1/-1 steps
    step = np.random.choice([-1,1], NSTEP)
    # calculate position of random walk at each step accumulating the individual values of step
    position = np.cumsum(step)
    plt.plot(position)
```
