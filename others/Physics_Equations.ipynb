### Imágenes con ecuaciones de física

```python
import numpy as np
import pylab as pl
```

```python
eqs = []

# Mechanics - Hamilton
eqs.append((r"$ \dot{\vec p_{i}} = - \frac{\partial H}{\partial \vec q_{i}} $"))
eqs.append((r"$ \dot{\vec q_{i}} = \frac{\partial H}{\partial \vec p_{i}} $"))
eqs.append((r"$ \frac{\partial H}{\partial t} = - \frac{\partial L}{\partial t} $"))

# Electrodynamics - Maxwell
eqs.append((r"$ \vec \nabla \cdot \vec E = \frac{\rho}{\epsilon_{0}} $"))
eqs.append((r"$ \vec \nabla \cdot \vec B = 0 $"))
eqs.append((r"$ \vec \nabla \times \vec E = - \frac{\partial \vec B}{\partial t} $"))
eqs.append((r"$ \vec \nabla \times \vec B = \mu_{0} \vec J + \mu_{0} \epsilon_{0} \frac{\partial \vec E}{\partial t} $"))

# Quantum
eqs.append((r"$i \hbar \frac{\mathrm{d}}{\mathrm{d}t} \left| \psi \right> = \hat H \left| \psi \right>$"))
eqs.append((r"$S = -tr(\rho \ln \rho)$"))

# General Relativity 
eqs.append((r"$R_{\mu\nu} - \frac{1}{2} g_{\mu\nu} R = \kappa T_{\mu\nu}$"))
eqs.append((r"$E = \sqrt{{m_0}^2c^4 + p^2c^2}$"))

pl.axes([0.025, 0.025, 3, 1])

for i in range(24):
    index = np.random.randint(0, len(eqs))
    eq = eqs[index]
    size = np.random.uniform(12, 32)
    x, y = np.random.uniform(0, 1, 2)
    alpha = np.random.uniform(0.25, .75)
    pl.text(
        x, 
        y, 
        eq, 
        ha='center', 
        va='center', 
        color="#11557c", 
        alpha=alpha, 
        transform=pl.gca().transAxes, 
        fontsize=size, 
        clip_on=True
    )

pl.xticks(())
pl.yticks(())
pl.show()
```
