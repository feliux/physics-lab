```python
import numpy as np              
import matplotlib.pyplot as plt 
```

```python
x = np.linspace(0,1,100)
y2 = np.sin(2*np.pi*x) * np.sqrt(2)
y3 = np.sin(3*np.pi*x) * np.sqrt(2)
y4 = np.zeros(100)
plt.plot(x, y2, label=r"$\psi_{2} = \sqrt{\frac{2}{L}}\sin(\frac{2\pi x}{L})$")
plt.plot(x, y3, label=r"$\psi_{3} = \sqrt{\frac{2}{L}}\sin(\frac{3\pi x}{L})$")
plt.plot(x, y4, "k")
plt.plot(2/3, 0, "ko")

nota = plt.annotate(r"$\frac{2L}{3}$",
         xy=(2/3, 0), xycoords="data",
         xytext=(2/3, -0.8), fontsize=20,
         arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

plt.xlabel(r"$L$")
#plt.title("Funciones de onda de una partícula en pozo de potencial infinito")
#plt.xticks([])
plt.yticks([])
plt.legend()
plt.show()
```

```python
x = np.linspace(0,1,100)
y2 = np.sin(2*np.pi*x) * np.sqrt(2)
y3 = np.sin(3*np.pi*x) * np.sqrt(2)
y4 = np.zeros(100)
plt.plot(x, y2, label=r"$\psi_{2}^{\prime} = \sqrt{\frac{2}{L^{\prime}}}\sin(\frac{2\pi x}{L^{\prime}})$")
plt.plot(x, y3, label=r"$\psi_{3}^{\prime} = \sqrt{\frac{2}{L^{\prime}}}\sin(\frac{3\pi x}{L^{\prime}})$")
plt.plot(x, y4, "k")
plt.plot(1/3, 0, "ko")
plt.plot(1/2, 0, "kd")
plt.plot(1, 0, "kd")

nota = plt.annotate(r"$x = \frac{2L}{3} = \frac{L^{\prime}}{3}$",
         xy=(1/3, 0), xycoords="data",
         xytext=(.04, -.5), fontsize=15,
         arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

nota = plt.annotate(r"$L$",
         xy=(1/2, 0), xycoords="data",
         xytext=(1/2, .5), fontsize=15,
         arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

nota = plt.annotate(r"$L^{\prime}$",
         xy=(1, 0), xycoords="data",
         xytext=(1, -.5), fontsize=15,
         arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

plt.xlabel(r"$L'$")
plt.yticks([])
plt.legend()
plt.show()
```

```python
x = np.linspace(0,1,100)
y2 = 2 * np.sin(2*np.pi*x) * np.sin(2*np.pi*x)
y3 = 2 * np.sin(3*np.pi*x) * np.sin(3*np.pi*x)
y4 = np.zeros(100)
plt.plot(x, y2, label=r"$|\psi_{2}|^{2}$")
plt.plot(x, y3, label=r"$|\psi_{3}|^{2}$")
plt.plot(x, y4, "k")
plt.plot(2/3, 0, "ko")

nota = plt.annotate(r"$\frac{2L}{3}$",
         xy=(2/3, 0), xycoords="data",
         xytext=(0.8, 0.5), fontsize=20,
         arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

plt.xlabel(r"$L$")
#plt.title("Funciones de onda de una partícula en pozo de potencial infinito")
#plt.xticks([])
plt.yticks([])
plt.legend()
plt.show()
```

```python
x1 = np.linspace(0,0.5,100)
x2 = np.linspace(0,1,100)
y = np.sin(np.pi*x2) * np.sqrt(2)
y2 = np.sin(2*np.pi*x2) * np.sqrt(2)
y3 = np.sin(3*np.pi*x2) * np.sqrt(2)
y4 = np.zeros(100)


ax = plt.subplot(111)

y1 =2 * np.sin(2*np.pi*x1)
ax.plot(x1, y1, label=r"$\psi_{1} = \sqrt{\frac{2}{L^{\prime}}}\sin(\frac{\pi x}{L^{\prime}})$")


ax.plot(x2, y, label=r"$\chi_{1} = \sqrt{\frac{2}{L}}\sin(\frac{\pi x}{L})$")
ax.plot(x2, y2, label=r"$\chi_{2} = \sqrt{\frac{2}{L}}\sin(\frac{2\pi x}{L})$")
ax.plot(x2, y3, label=r"$\chi_{3} = \sqrt{\frac{2}{L}}\sin(\frac{3\pi x}{L})$")
ax.plot(x2, y4, "k")


ax.legend(loc="center left", bbox_to_anchor=(1, 0.5))


nota = plt.annotate(r"$\frac{L}{2}$",
         xy=(1/2, 0), xycoords="data",
         xytext=(0.55, .4), fontsize=15,
         arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

plt.axvline(x=0.5, color="k")
plt.axvline(x=1, color="k")
plt.axvline(x=0, color="k")

plt.xlabel(r"$L$")
plt.yticks([])
plt.show()
```

```python
t = np.arange(0.1, 20, 0.1)

y1 = np.sin(t)/t
y2 = np.sin(t)*np.exp(-t)
p1, p2 = plt.plot(t, y1, t, y2)

# Texto en la gráfica en coordenadas (x,y)
texto1 = plt.text(2, 0.6, r"$\frac{\sin(x)}{x}$", fontsize=20)
texto2 = plt.text(13, 0.2, r"$\sin(x) \cdot e^{-x}$", fontsize=16)

# Añado una malla al gráfico
plt.grid()

plt.title("Representacion de dos funciones")
plt.xlabel("Tiempo / s")
plt.ylabel("Amplitud / cm")


# Punto a señalar en la primera gráfica
px = 7.5
py = np.sin(px)/px

# Pinto las coordenadas con un punto negro
punto = plt.plot([px], [py], "bo")

# Hago un señalización con flecha
nota = plt.annotate(r"$\frac{\sin(7.5)}{\exp(-7.5)} = 0.12$",
         xy=(px, py), xycoords="data",
         xytext=(3, 0.4), fontsize=9,
         arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))
```

```python
def f1(x):
    y = np.sin(x)
    return y

def f2(x):
    y = np.sin(x) + np.sin(5.0*x)
    return y


def f3(x):
    y = np.sin(x) * np.exp(-x/10.)
    return y

# array de valores que quiero representar
x = np.linspace(0, 10*np.pi, 800)

# Creo una figura (ventana), pero indico el tamaño (x,y) en pulgadas
plt.figure(figsize=(12, 5))

p1, p2, p3 = plt.plot(x, f1(x), x, f2(x), x, f3(x))

# Añado leyenda, tamaño de letra 10, en esquina superior derecha
plt.legend(("Funcion 1", "Funcion 2", "Funcion 3"), prop={"size":10}, loc="upper right")

plt.xlabel("Tiempo / s")
plt.ylabel("Amplitud / cm")
plt.title("Representacion de tres funciones")

# Plots con label
p1 = plt.plot(x, f1(x), label="Funcion 1")
p2 = plt.plot(x, f2(x), label="Funcion 2")
p3 = plt.plot(x, f3(x), label="Funcion 3")

# Ahora se puede usar legend sin etiqueta, pero indico que aparezca abajo derecha
plt.legend(loc="lower right")
```
